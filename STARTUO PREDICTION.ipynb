```python
import pandas as pd

data  = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/startupdata/training_set_label.csv" )

data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data['state_code'] = encoder.fit_transform(data['state_code'])
data['status'] = encoder.fit_transform(data['status'])
data['city'] = encoder.fit_transform(data['city'])
data['name'] = encoder.fit_transform(data['name'])
data['is_mobile'] = encoder.fit_transform(data['is_mobile'])
data.head()
```

```python
data.fillna(786)
data.head()
```

```python
"""
'Unnamed: 0', 'state_code', 'latitude', 'longitude', 'zip_code', 'id',
       'city', 'Unnamed: 6', 'name', 'founded_at', 'closed_at',
       'first_funding_at', 'last_funding_at', 'age_first_funding_year',
       'age_last_funding_year', 'age_first_milestone_year',
       'age_last_milestone_year', 'relationships', 'funding_rounds',
       'funding_total_usd', 'milestones', 'state_code.1', 'is_CA', 'is_NY',
       'is_MA', 'is_TX', 'is_otherstate', 'category_code', 'is_software',
       'is_web', 'is_mobile', 'is_enterprise', 'is_advertising',
       'is_gamesvideo', 'is_ecommerce', 'is_biotech', 'is_consulting',
       'is_othercategory', 'object_id', 'has_VC', 'has_angel', 'has_roundA',
       'has_roundB', 'has_roundC', 'has_roundD', 'avg_participants',
       'is_top500', 'status'
"""

data['category_code'] = encoder.fit_transform(data['category_code'])
```

```python
"""
'Unnamed: 0', 'state_code', 'latitude', 'longitude', 'zip_code', 'id',
       'city', 'Unnamed: 6', 'name', 'founded_at', 'closed_at',
       'first_funding_at', 'last_funding_at', 'age_first_funding_year',
       'age_last_funding_year', 'age_first_milestone_year',
       'age_last_milestone_year', 'relationships', 'funding_rounds',
       'funding_total_usd', 'milestones', 'state_code.1', 'is_CA', 'is_NY',
       'is_MA', 'is_TX', 'is_otherstate', 'category_code', 'is_software',
       'is_web', 'is_mobile', 'is_enterprise', 'is_advertising',
       'is_gamesvideo', 'is_ecommerce', 'is_biotech', 'is_consulting',
       'is_othercategory', 'object_id', 'has_VC', 'has_angel', 'has_roundA',
       'has_roundB', 'has_roundC', 'has_roundD', 'avg_participants',
       'is_top500', 'status'
"""
X = data.drop(['Unnamed: 0','Unnamed: 6', 'status', 'object_id', 'id', 'founded_at', 'closed_at',
       'first_funding_at', 'last_funding_at', 'state_code.1', 'zip_code'], axis=1)
#drop([' shares'], axis=1)
Y = data['status']

xData = X.values
yData = Y.values
```

```python
from sklearn.model_selection import train_test_split 

xTrain, xTest, yTrain, yTest = train_test_split( 
        xData, yData, test_size = 0.1, random_state = 365) 
```

```python
data[data.eq('CA 94105').any(1)]
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
model = XGBClassifier()
model.fit(xTrain, yTrain)
y_pred=model.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)
```

```python
test = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/startupdata/testing_set_label.csv')

test['state_code'] = encoder.fit_transform(test['state_code'])
test['city'] = encoder.fit_transform(test['city'])
test['name'] = encoder.fit_transform(test['name'])
test['is_mobile'] = encoder.fit_transform(test['is_mobile'])
test['category_code'] = encoder.fit_transform(test['category_code'])

cols_when_model_builds = model.get_booster().feature_names
cols_when_model_builds
```

```python
X = test.drop(['Unnamed: 0','Unnamed: 6', 'object_id', 'id', 'founded_at', 'closed_at',
       'first_funding_at', 'last_funding_at', 'state_code.1', 'zip_code'], axis=1)
vals = model.predict(X.values)
Vals = []
for i in vals:
    if i == 0: Vals.append('acquired')
    else: Vals.append('closed')
```

```python
res = pd.DataFrame(Vals)
res.columns = ["prediction"]
res.to_csv("StartUpPred.csv", index = False)
```
