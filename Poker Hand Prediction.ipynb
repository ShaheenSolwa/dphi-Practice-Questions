```python
import pandas as pd

data = pd.read_csv("C:/Users/shahe/Desktop/Train_data.csv" )

data.head()
```

```python
data = data.fillna('786')
```

```python
X = data.drop(['Class'], axis=1)
Y = data['Class']

xData = X.values
yData = Y.values
```

```python
from sklearn.model_selection import train_test_split 

xTrain, xTest, yTrain, yTest = train_test_split( 
        xData, yData, test_size = 0.05, random_state=42) 
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, f1_score
model = XGBClassifier()
model.fit(xTrain, yTrain)
y_pred=model.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)
print(f1_score(yTest, y_pred, average=None))
```

```python
test_data = pd.read_csv('C:/Users/shahe/Desktop/Test_data.csv')
test_data.head()
```

```python
vals = model.predict(test_data.values)
```

```python
res = pd.DataFrame(vals)
res.columns = ["prediction"]
res.to_csv("Poker_pred.csv", index = False)
```
