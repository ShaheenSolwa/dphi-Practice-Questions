```python
import pandas as pd

data  = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/bank_marketing_data/training_set_label.csv" )

data.head()
```

```python
list(set(data.job))
data['job'] = data['job'].map({
    'blue-collar':0,
 'management':1,
 'self-employed':2,
 'services':3,
 'unemployed':4,
 'entrepreneur':5,
 'technician':6,
 'retired':7,
 'admin.':8,
 'student':9,
 'housemaid':10,
 'unknown':11
})
data.head()
```

```python
list(set(data.marital))
data['martial'] = data['marital'].map({
    'married':0, 
    'single':1, 
    'divorced':2
})
data.head()
```

```python
list(set(data['education']))
data['education'] = data['education'].map({
    'tertiary':0, 
    'primary':1, 
    'secondary':2, 
    'unknown':3
})
data.head()
```

```python
list(set(data['default']))
data['default'] = data['default'].map({
    'yes':0,
    'no':1
})
data.head()
```

```python
list(set(data['housing']))
data['housing'] = data['housing'].map({
    'yes':0,
    'no':1
})
data.head()
```

```python
list(set(data['loan']))
data['loan'] = data['loan'].map({
    'yes':0,
    'no':1
})
data.head()
```

```python
list(set(data['month']))
data['month'] = data['month'].map({
    'nov':10,
 'feb':1,
 'oct':9,
 'mar':2,
 'jul':6,
 'jun':5,
 'apr':3,
 'jan':0,
 'may':4,
 'dec':11,
 'sep':8,
 'aug':7
})
data.head()
```

```python
list(set(data['marital']))
data['marital'] = data['marital'].map({
    'married':0, 
    'single':1, 
    'divorced':3
})
data.head
```

```python
list(set(data['contact']))
data['contact'] = data['contact'].map({
    'telephone':0, 'cellular':1, 'unknown':2
})
```

```python
list(set(data['poutcome']))
data['poutcome'] = data['poutcome'].map({
    'failure':0, 'success':1, 'unknown':2, 'other':3
})
```

```python
data.head()
```

```python
data.drop(['martial'], inplace=True, axis=1)
data.head()
```

```python
X = data.drop(['subscribe'], axis=1)
Y = data.subscribe

xData = X.values
yData = Y.values
```

```python
from sklearn.model_selection import train_test_split 

xTrain, xTest, yTrain, yTest = train_test_split( 
        xData, yData, test_size = 0.1, random_state = 365) 
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
model = XGBClassifier()
model.fit(xTrain, yTrain)
y_pred=model.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)
```
