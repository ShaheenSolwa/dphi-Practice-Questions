```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/breast_cancer/Training_set_breastcancer.csv')
data.head()
```

```python
data['diagnosis'] = data['diagnosis'].map({
    'B': 0,
    'M': 1
})

data.head()
```

```python
X = data.drop(['id', 'diagnosis'], axis=1)
Y = data.diagnosis

xData = X.values
yData = Y.values
```

```python
from sklearn.model_selection import train_test_split 

xTrain, xTest, yTrain, yTest = train_test_split( 
        xData, yData, test_size = 0.33, random_state = 365) 
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
model = XGBClassifier()
model.fit(xTrain, yTrain)
y_pred=model.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=25, random_state=365)
clf.fit(xTrain, yTrain)
y_pred=clf.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)
```

```python
Eval = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/breast_cancer/Testing_set_breastcancer.csv')
```

```python
test = Eval.drop(['id'],axis=1)
test = test.values
vals = model.predict(test)
```

```python
Vals = []
for i in vals:
    if i == 0: Vals.append('B')
    else: Vals.append('M')
```

```python
res = pd.DataFrame(Vals)
res.columns = ["prediction"]
res.to_csv("BCancePred.csv", index = False)
```
