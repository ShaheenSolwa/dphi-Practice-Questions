```python
import pandas as pd

data = pd.read_csv(r'C:/Users/shahe/Desktop/tictactoe.csv')

data.head()
```

```python
data['result'] = data['result'].map({
    'positive': 0,
    'negative': 1
})

data.head()
```

```python
data['tls'] = data['tls'].map({
    'x': 0,
    'o': 1,
    'b': 2
})
data['tms'] = data['tms'].map({
    'x': 0,
    'o': 1,
    'b': 2
})
data['trs'] = data['trs'].map({
    'x': 0,
    'o': 1,
    'b': 2
})
data['mls'] = data['mls'].map({
    'x': 0,
    'o': 1,
    'b': 2
})
#mms	mrs	bls	bms	brs
data['mms'] = data['mms'].map({
    'x': 0,
    'o': 1,
    'b': 2
})
data['mrs'] = data['mrs'].map({
    'x': 0,
    'o': 1,
    'b': 2
})
data['bls'] = data['bls'].map({
    'x': 0,
    'o': 1,
    'b': 2
})
data['bms'] = data['bms'].map({
    'x': 0,
    'o': 1,
    'b': 2
})
data['brs'] = data['brs'].map({
    'x': 0,
    'o': 1,
    'b': 2
})

data.head()
```

```python
X = data.drop(['result'], axis=1)
Y = data.result

xData = X.values
yData =Y.values
```

```python
from sklearn.model_selection import train_test_split 

xTrain, xTest, yTrain, yTest = train_test_split( 
        xData, yData, test_size = 0.1, random_state = 42) 
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(xTrain,yTrain)
y_pred=logreg.predict(xTest)

from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(yTest, y_pred))
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
model = XGBClassifier()
model.fit(xTrain, yTrain)
y_pred=model.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)
```
