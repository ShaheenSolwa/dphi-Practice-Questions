```python
import pandas as pd

data = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/crop_recommendation/train_set_label.csv")
```

```python
data.head()
```

```python
data['crop'] = data['crop'].map({
    'muskmelon':0,
 'pomegranate':1,
 'maize':2,
 'grapes':3,
 'rice':4,
 'mango':5,
 'mothbeans':6,
 'pigeonpeas':7,
 'kidneybeans':8,
 'jute':9,
 'mungbean':10,
 'orange':11,
 'blackgram':12,
 'apple':13,
 'papaya':14,
 'lentil':15,
 'cotton':16,
 'banana':17,
 'coconut':18,
 'coffee':19,
 'chickpea':20,
 'watermelon':21
})
data.head()
```

```python
X = data.drop(['crop'], axis=1)
Y = data['crop']

print(X.shape)
print(Y.shape)

xData = X.values
yData = Y.values
```

```python
from sklearn.model_selection import train_test_split 

xTrain, xTest, yTrain, yTest = train_test_split( 
        xData, yData, test_size = 0.2, random_state = 0) 
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
model = XGBClassifier()
model.fit(xTrain, yTrain)
y_pred=model.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=25, random_state=0)
clf.fit(xTrain, yTrain)
y_pred=clf.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)
```

```python
test_data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/crop_recommendation/test_set_label.csv')
```

```python
vals = clf.predict(test_data.values)
```

```python
preds = []
for i in vals:
    if i == 0: preds.append('muskmelon')
    elif i == 1: preds.append('pomegranate')
    elif i == 2: preds.append('maize')
    elif i == 3: preds.append('grapes')
    elif i == 4: preds.append('rice')
    elif i == 5: preds.append('mango')
    elif i == 6: preds.append('mothbeans')
    elif i == 7: preds.append('pigeonpeas')
    elif i == 8: preds.append('kidneybeans')
    elif i == 9: preds.append('jute')
    elif i == 10: preds.append('mungbean')
    elif i == 11: preds.append('orange')
    elif i == 12: preds.append('blackgram')
    elif i == 13: preds.append('apple')
    elif i == 14: preds.append('papaya')
    elif i == 15: preds.append('lentil')
    elif i == 16: preds.append('cotton')
    elif i == 17: preds.append('banana')
    elif i == 18: preds.append('coconut')
    elif i == 19: preds.append('coffee')
    elif i == 20: preds.append('chickpea')
    elif i == 21: preds.append('watermelon')
```

```python
res = pd.DataFrame(preds)
res.columns = ["prediction"]
res.to_csv("CropPred.csv", index = False)
```
