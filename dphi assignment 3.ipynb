```python
import pandas as pd

data  = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/cancer_death_rate/Training_set_label.csv" )

data.head()
```

```python
data.columns
```

```python
Y = data.TARGET_deathRate
X = data.drop(['TARGET_deathRate', 'binnedInc', 'Geography'], axis=1)

xData = X.values
yData = Y.values
```

```python
from sklearn.model_selection import train_test_split 

xTrain, xTest, yTrain, yTest = train_test_split( 
        xData, yData, test_size = 0.33, random_state = 365) 
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
model = XGBClassifier()
model.fit(xTrain, yTrain)
y_pred=model.predict(xTest)
```

```python
_data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/cancer_death_rate/Testing_set_label.csv')

X = _data.drop(['binnedInc', 'Geography'], axis=1)

vals = model.predict(X.values)

res = pd.DataFrame(vals)
res.columns = ["prediction"]
res.to_csv("Shaheen_Solwa_A3_Beg.csv", index = False)
```
