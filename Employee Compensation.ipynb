```python
import pandas as pd

data = pd.read_csv("C:/Users/shahe/Desktop/train_set.csv")

data.head()
```

```python
data = data.fillna('786')
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

data['OG'] = le.fit_transform(data['OG'])
data['DC'] = le.fit_transform(data['DC'])
data['Union'] = le.fit_transform(data['Union'])
data['JF'] = le.fit_transform(data['JF'])
data['Job'] = le.fit_transform(data['Job'])
data['YT'] = le.fit_transform(data['YT'])

data.head()
```

```python
data['Dept'] = le.fit_transform(data['Dept'])

data.head()
```

```python
X = data.drop(['Total_Compensation','Dept','Union','EI','YT'], axis = 1)
y = data['Total_Compensation']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X,y,test_size=0.1, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
pred = lr.predict(X_val)

from sklearn.metrics import mean_squared_error
import numpy as np
print('Root Mean Squared Error is: ', np.sqrt(mean_squared_error(y_val, pred)))
```

```python
test = test.fillna('786')
test['OG'] = le.fit_transform(test['OG'])
test['DC'] = le.fit_transform(test['DC'])
test['Union'] = le.fit_transform(test['Union'])
test['JF'] = le.fit_transform(test['JF'])
test['Job'] = le.fit_transform(test['Job'])
test['YT'] = le.fit_transform(test['YT'])

test = test.drop(['Dept','Union','EI','YT'], axis = 1)

target = lr.predict(test.values)
```
