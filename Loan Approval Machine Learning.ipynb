```python
import pandas as pd

data = pd.read_csv(r'C:/Users/shahe/Desktop/LoadApproval.csv')

data.head()
```

```python
data = data.drop(['Loan_ID'], axis=1)
```

```python
data['Gender'] = data['Gender'].map({'Female': int(1), 'Male': int(0)})
data.head()
```

```python
data['Married'] = data['Married'].map({'Yes': int(1), 'No': int(0)})
data.head()
```

```python
data['Education'] = data['Education'].map({'Graduate': int(1), 'Not Graduate': int(0)})
data.head()
```

```python
data['Self_Employed'] = data['Self_Employed'].map({'Yes': int(1), 'No': int(0)})
data.head()
```

```python
data['ApplicantIncome'] = data['ApplicantIncome'].fillna(0)
data['CoapplicantIncome'] = data['CoapplicantIncome'].fillna(0)
data['LoanAmount'] = data['LoanAmount'].fillna(0)

data.head()
```

```python
data['Property_Area'] = data['Property_Area'].map({'Urban': int(1), 'Rural': int(0)})
data.head()
```

```python
data['Loan_Status'] = data['Loan_Status'].map({'Y': int(1), 'N': int(0)})
data.head()
```

```python
data['Dependents'] = data['Dependents'].map({'0': int(0),'1': int(1), '2': int(2), '3+': int(3)})
data.head()
```

```python
X = data.drop(['Loan_Status'], axis=1)
Y = data['Loan_Status']

xData = X.values
yData = Y.values
```

```python
from sklearn.model_selection import train_test_split 
# Split the data into training and testing sets 
xTrain, xTest, yTrain, yTest = train_test_split( 
        xData, yData, test_size = 0.2, random_state = 42) 
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
model = XGBClassifier()
model.fit(xTrain, yTrain)
y_pred=model.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)
```
