```python
import pandas as pd

data  = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/eng_grad_emp_salary/training_set_label.csv" )

data.head()
```

```python
data['Gender'] = data['Gender'].map({
    'f': 0,
    'm': 1
})
data.head()
```

```python
data.columns
```

```python
list(set(data['CollegeState']))
data['CollegeState'] = data['CollegeState'].map({
    'Jharkhand':0,
 'Uttarakhand':1,
 'Haryana':2,
 'Jammu and Kashmir':3,
 'Sikkim':4,
 'Maharashtra':5,
 'Tamil Nadu':6,
 'Himachal Pradesh':7,
 'Meghalaya':8,
 'Orissa':9,
 'Bihar':10,
 'Madhya Pradesh':11,
 'Assam':12,
 'Delhi':13,
 'Rajasthan':14,
 'Punjab':15,
 'Union Territory':16,
 'Kerala':17,
 'Uttar Pradesh':18,
 'Gujarat':19,
 'Goa':20,
 'Andhra Pradesh':21,
 'Chhattisgarh':22,
 'West Bengal':23,
 'Karnataka':24,
 'Telangana':25
})
data.head()
```

```python
list(set(data['Degree']))
data['Degree'] = data['Degree'].map({
    'MCA':0, 
    'M.Tech./M.E.':1, 
    'B.Tech/B.E.':2, 
    'M.Sc. (Tech.)':3
})
data.head()
```

```python
X = data[['Gender', '10percentage', '12percentage', 'collegeGPA',
         'GraduationYear', 'English', 'Logical', 'Quant',
       'Domain', 'ComputerProgramming', 'ElectronicsAndSemicon',
       'ComputerScience', 'MechanicalEngg', 'ElectricalEngg', 'TelecomEngg',
       'CivilEngg', 'conscientiousness', 'agreeableness', 'extraversion',
       'nueroticism', 'openess_to_experience']]
Y = data.Salary

xData = X.values
yData = Y.values
```

```python
from sklearn.model_selection import train_test_split 

xTrain, xTest, yTrain, yTest = train_test_split( 
        xData, yData, test_size = 0.33, random_state = 365) 
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(xTrain,yTrain)
y_pred=logreg.predict(xTest)

from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(yTest, y_pred))
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
model = XGBClassifier()
model.fit(xTrain, yTrain)
y_pred=model.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=25, random_state=365)
clf.fit(xTrain, yTrain)
y_pred=model.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)
```
