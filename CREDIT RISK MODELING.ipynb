```python
import pandas as pd

data  = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/credit_risk/training_set_labels.csv" )

data.head()
```

```python
list(set(data.own_telephone))
data['own_telephone'] = data['own_telephone'].map({
    'yes': 0,
    'none': 1
})
```

```python
list(set(data.other_parties))
data['other_parties'] = data['other_parties'].map({
    'co applicant': 0,
    'none': 1,
    'guarantor': 2
})
data.head()
```

```python
list(set(data.personal_status))
data['personal_status'] = data['personal_status'].map({
    "'male div/sep'": 0,
    "'female div/dep/mar'": 1,
    "'male mar/wid'": 2,
    "'male single'": 3
})
data.head()
```

```python
list(set(data.foreign_worker))
data['foreign_worker'] = data['foreign_worker'].map({
    'yes': 0,
    'no': 1,
})
data.head()
```

```python
list(set(data.installment_commitment))
```

```python
list(set(data.other_payment_plans))
data['other_payment_plans'] = data['other_payment_plans'].map({
    'stores': 0,
    'none': 1,
    'bank': 2
})
data.head()
```

```python
list(set(data.checking_status))
data['checking_status'] = data['checking_status'].map({
    "'no checking'": 0,
    "'<0'": 1,
    "'>=200'": 2,
    "'0<=X<200'": 3
})
data.head()
```

```python
list(set(data.duration))
```

```python
list(set(data.employment))
data['employment'] = data['employment'].map({
    "'1<=X<4'": 0,
    "'>=7'": 1,
    "'4<=X<7'": 2,
    "'<1'": 3
})
data.head()
```

```python
list(set(data.residence_since))
```

```python
list(set(data['credit_amount']))
data.head()
```

```python
list(set(data.job))
data['job'] = data['job'].map({
    "'high qualif/self emp/mgmt'": 0,
    "'umemp/unskilled non res'": 1,
    'skilled': 2,
    "'unskilled resident'": 3
})
data.head()
```

```python
list(set(data.num_dependents))
```

```python
list(set(data.age))
```

```python
list(set(data.savings_status))
data['savings_status'] = data['savings_status'].map({
    "'100<=X<500'": 0,
    "'>=1000'": 1,
    "'500<=X<1000'": 2,
    "'<100'": 3,
    "'no known savings'": 4
})
data.head()
```

```python
list(set(data.housing))
data['housing'] = data['housing'].map({
    "'for free'": 0,
    'rent': 1,
    'own': 2
})
data.head()
```

```python
list(set(data.property_magnitude))
data['property_magnitude'] = data['property_magnitude'].map({
    'car': 0,
    "'life insurance'": 1,
    "'no known property'": 2,
    "'real estate'": 3
})
data.head()
```

```python
list(set(data.existing_credits))
```

```python
list(set(data.purpose))
data['purpose'] = data['purpose'].map({
    'other': 0,
    'education': 1,
    'furniture/equipment': 2,
    'radio/tv': 3,
    "'domestic appliance'": 4,
    'repairs': 5,
    'retraining': 6,
    "'used car'": 7,
    'business': 8,
    "'new car'": 9
})
data.head()
```

```python
list(set(data.credit_history))
data['credit_history'] = data['credit_history'].map({
    "'critical/other existing credit'": 0,
    "'existing paid'": 1,
    "'no credits/all paid'": 2,
    "'all paid'": 3,
    "'delayed previously'": 4
})
data.head()
```

```python
list(set(data['class']))
data['class'] = data['class'].map({
    'good': 0,
    'bad': 1
})
data.head()
```

```python
data.head()
```

```python
data = data.fillna(0)
```

```python
data.head()
```

```python
Y = data['class']
X = data.drop(['class'], axis=1)

xData = X.values
yData = Y.values
```

```python
from sklearn.model_selection import train_test_split 

xTrain, xTest, yTrain, yTest = train_test_split( 
        xData, yData, test_size = 0.33, random_state = 365) 
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
model = XGBClassifier()
model.fit(xTrain, yTrain)
y_pred=model.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=25, random_state=365)
clf.fit(xTrain, yTrain)
y_pred=clf.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)
```
