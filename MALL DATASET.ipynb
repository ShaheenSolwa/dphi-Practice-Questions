```python
import pandas as pd

data = pd.read_csv(r"C:/Users/shahe/Desktop/MallCustomers.csv" )

data.head()
```

```python
data['Gender'] = data['Gender'].map({'Male': 1, 'Female': 0})

data.head()
```

```python
X = data.drop(['CustomerID', 'Spending_Score'], axis=1)
Y = data['Spending_Score']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, 
                                test_size=0.33, 
                                random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(x_train,y_train)
y_pred=logreg.predict(x_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
model = XGBClassifier()
model.fit(x_train, y_train)
y_pred=model.predict(x_test)
print(accuracy_score(y_test, y_pred)*100)
```

```python
Dataset = data.drop(['CustomerID'], axis=1).values
X = Dataset[:, 0:3]
Y = Dataset[:, 3]
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)

X_scale
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, Y_train, Y_val = train_test_split(X_scale, Y, test_size=0.5)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential([    Dense(32, activation='relu', input_shape=(3,)),    Dense(32, activation='relu'),    Dense(1, activation='sigmoid'),])
model.compile(optimizer='adam',              loss='binary_crossentropy',              metrics=['accuracy'])
hist = model.fit(X_train, Y_train,          batch_size=32, epochs=150,          validation_data=(X_val, Y_val))
```

```python
_, accuracy = model.evaluate(X_val, Y_val)
print('Accuracy: %.2f' % (accuracy*100))
```
