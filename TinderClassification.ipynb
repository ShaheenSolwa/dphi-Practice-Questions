```python
import pandas as pd

data = pd.read_csv("C:/Users/shahe/Desktop/train_set_label.csv" )

data.head()
```

```python
data['Segment type'] = data['Segment type'].map({
   'gender':0, 
   'mobile':1,
   'university':2, 
   'web':3 
})
data['Answer'] = data['Answer'].map({
   'No':0, "I don't use Tinder":1, 'Yes':2
})
```

```python
data.head()
```

```python
urls = data['Segment Description']
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data['Segment Description'] = encoder.fit_transform(urls)
data.head()
```

```python
X = data.drop(['It became a relationship'], axis=1)
Y = data['It became a relationship']

print(X.shape)
print(Y.shape)

xData = X.values
yData = Y.values
```

```python
from sklearn.model_selection import train_test_split 

xTrain, xTest, yTrain, yTest = train_test_split( 
        xData, yData, test_size = 0.1) 
```

```python
from sklearn.model_selection import train_test_split 
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

xTrain, xTest, yTrain, yTest = train_test_split( 
        xData, yData, test_size=0.1) 


model = XGBClassifier()
model.fit(xTrain, yTrain)
y_pred=model.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)

while (accuracy_score(yTest, y_pred)*100) <= 94.0:
    xTrain, xTest, yTrain, yTest = train_test_split( 
        xData, yData) 
    model = XGBClassifier()
    model.fit(xTrain, yTrain)
    y_pred=model.predict(xTest)
    print(accuracy_score(yTest, y_pred)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=10, random_state=0)
clf.fit(xTrain, yTrain)
y_pred=clf.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)
```
