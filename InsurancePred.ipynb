```python
import pandas as pd

data = pd.read_csv('C:/Users/shahe/Desktop/insurance_train.csv')

data.head()
```

```python
data = data.fillna('786')
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

le.fit(data['Agency'])
data['Agency'] = le.transform(data['Agency'])
```

```python
data.head()
```

```python
le.fit(data['Agency Type'])
data['Agency Type'] = le.transform(data['Agency Type'])

le.fit(data['Distribution Channel'])
data['Distribution Channel'] = le.transform(data['Distribution Channel'])

le.fit(data['Product Name'])
data['Product Name'] = le.transform(data['Product Name'])

le.fit(data['Destination'])
data['Destination'] = le.transform(data['Destination'])

le.fit(data['Destination'])
data['Destination'] = le.transform(data['Destination'])

le.fit(data['Gender'])
data['Gender'] = le.transform(data['Gender'])
```

```python
X = data.drop(['Claim'], axis=1)
Y = data.Claim

xData = X.values
yData = Y.values
```

```python
from sklearn.model_selection import train_test_split 

xTrain, xTest, yTrain, yTest = train_test_split( 
        xData, yData, test_size = 0.1, random_state = 42) 
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(xTrain,yTrain)
y_pred=logreg.predict(xTest)

from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(yTest, y_pred))
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
model = XGBClassifier()
model.fit(xTrain, yTrain)
y_pred=model.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)
```
