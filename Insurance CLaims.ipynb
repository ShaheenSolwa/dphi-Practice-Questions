```python
import pandas as pd

data = pd.read_csv("C:/Users/shahe/Desktop/train_data.csv")

data.head()
```

```python
data["Building_Painted"] = data["Building_Painted"].map({
    'N': 0,
    'V': 1
})
```

```python
data["Building_Fenced"] = data["Building_Fenced"].map({
    'N': 0,
    'V': 1
})
```

```python
data["Garden"] = data["Garden"].map({
    'O': 0,
    'V': 1
})
```

```python
data["Settlement"] = data["Settlement"].map({
    'R': 0,
    'U': 1
})
```

```python
data = data.fillna(786)
data.head()
```

```python
data["NumberOfWindows"] = data["NumberOfWindows"].map({
    '6':6, '7':7, '4':4, '9':9, '8':8, '   .':0, '2':2, '>=10':10, '5':5, '3':3, '1':1
})
```

```python
data.head()
```

```python
data = data.select_dtypes(exclude='object')
```

```python
data.head()
```

```python
X = data.drop(['Claim'], axis=1)
Y = data['Claim']

xData = X.values
yData = Y.values
```

```python
from sklearn.model_selection import train_test_split 

xTrain, xTest, yTrain, yTest = train_test_split( 
        xData, yData, test_size = 0.33, random_state = 42) 
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
model = XGBClassifier()
model.fit(xTrain, yTrain)
y_pred=model.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(xTrain,yTrain)
y_pred=logreg.predict(xTest)

from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(yTest, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=100, random_state=42)
clf.fit(xTrain, yTrain)
y_pred=clf.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)
```
