```python
import pandas as pd

data = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/used_car_price/train_set_label.csv")

data.head()
```

```python
data.columns
```

```python
data = data.select_dtypes(exclude=['object'])
data.head()
```

```python
data = data.fillna('786')
data.head()
```

```python
X = data.drop(['price_of_car'], axis=1)
Y = data['price_of_car']

xData = X.values
yData = Y.values

from sklearn.model_selection import train_test_split 

xTrain, xTest, yTrain, yTest = train_test_split( 
        xData, yData, test_size = 0.2) 
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
model = XGBClassifier()
model.fit(xTrain, yTrain)
y_pred=model.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(xTrain,yTrain)
y_pred=logreg.predict(xTest)

from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(yTest, y_pred))
```

```python
test = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/used_car_price/test_set_label.csv')
```

```python
test = test.select_dtypes(exclude=['object'])
test = test.fillna('786')
```

```python
vals = model.predict(test.values)
```
