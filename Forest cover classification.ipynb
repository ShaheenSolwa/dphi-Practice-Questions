```python
import pandas as pd

mo_data = pd.read_csv("Train_data (2).csv" )

mo_data.head()
```

```python
mo_data.fillna('786')
```

```python
X = mo_data.drop(['Cover_Type'], axis=1)
Y = mo_data['Cover_Type']

print(X.shape)
print(Y.shape)

xData = X.values
yData = Y.values
```

```python
from sklearn.model_selection import train_test_split 

xTrain, xTest, yTrain, yTest = train_test_split( 
        xData, yData, test_size = 0.01, random_state = 0) 
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
model = XGBClassifier()
model.fit(xTrain, yTrain)
y_pred=model.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
clf = RandomForestClassifier(max_depth=50)
clf.fit(xTrain, yTrain)
y_pred = clf.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)
```

```python
test_data = pd.read_csv('Test_data (2).csv')
test_data.head()
```

```python
print(len(test_data))
vals = clf.predict(test_data.values)
```

```python
res = pd.DataFrame(vals)
res.columns = ["prediction"]
res.to_csv("Forrest_Cover_preds.csv", index = False)
```
