```python
import pandas as pd

data  = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/abalone_data/training_set_label.csv" )

data.head()
```

```python
data['Sex'] = data['Sex'].map({
    'F': 0,
    'M': 1
})

data.head()
```

```python
X = data.drop(['Rings'], axis=1)
Y = data.Rings

xData = X.values
yData = Y.values
```

```python
from sklearn.model_selection import train_test_split 

xTrain, xTest, yTrain, yTest = train_test_split( 
        xData, yData, test_size = 0.1, random_state = 365) 
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
model = XGBClassifier()
model.fit(xTrain, yTrain)
y_pred=model.predict(xTest)
print(accuracy_score(yTest, y_pred)*100)
```

```python
test = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/abalone_data/testing_set_label.csv')
test['Sex'] = test['Sex'].map({
    'F': 0,
    'M': 1
})
test.head()
```

```python
vals = model.predict(test.values)
```

```python
res = pd.DataFrame(vals)
res.columns = ["prediction"]
res.to_csv("AbalonePred.csv", index = False)
```
