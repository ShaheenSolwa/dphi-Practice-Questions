```python
import pandas as pd
data  = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/bike_data/bike_train.csv" )

data.head()
```

```python
list(set(data['dteday']))
```

```python
X = data.drop(['cnt', 'dteday'],axis=1)
Y = data.cnt

xData = X.values
yData = Y.values
```

```python
from sklearn.model_selection import train_test_split 

xTrain, xTest, yTrain, yTest = train_test_split( 
        xData, yData, test_size = 0.33, random_state = 0) 
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(xTrain,yTrain)
y_pred=logreg.predict(xTest)

from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(yTest, y_pred))
```

```python
test_data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/bike_data/bike_test.csv')

x = test_data.drop(['dteday'], axis=1)

c = x.values
```

```python
vals = logreg.predict(c)
```

```python
res = pd.DataFrame(vals)
res.columns = ["prediction"]
res.to_csv("BikePred.csv", index = False)
```
