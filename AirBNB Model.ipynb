```python
import pandas as pd
airbnb_data = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/airbnb%20_data/airbnb_listing_train.csv" )

airbnb_data.head()
```

```python
list(set(airbnb_data.room_type))
```

```python
airbnb_data['room_type'] = airbnb_data['room_type'].map({
    'Private room': 0,
    'Entire home/apt': 1,
    'Hotel room': 2,
    'Shared room': 3
})

airbnb_data.head()
```

```python
X = airbnb_data[['latitude','longitude','room_type', 'minimum_nights', 'availability_365']]
Y = airbnb_data['price']

xData = X.values
yData = Y.values
```

```python
from sklearn.model_selection import train_test_split 

xTrain, xTest, yTrain, yTest = train_test_split( 
        xData, yData, test_size = 0.33, random_state = 45) 
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(xTrain,yTrain)
y_pred=logreg.predict(xTest)

from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(yTest, y_pred))
```
